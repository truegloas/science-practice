```python
from abc import abstractmethod

import numpy as np
import pandas as pd
import time
```

```python
class Hash_Table(dict):
    
    def __init__(self, size=1000):
        self.table = {}
        self.size = size
        for i in range(size):
            self.table.update({i : None})
            pass
        pass
    
    @abstractmethod
    def find(self, key):
        pass
    
    @abstractmethod
    def insert(self, key):
        pass
    
    @abstractmethod
    def delete(self, key):
        pass
    
    @abstractmethod
    def class_name(self):
        pass
```

```python
class Hash_Remainder_Division(Hash_Table):
    
    def hash_key(self, key, radix=10):
        
        key = list(bytes(key, 'utf-8'))
        
        buff = 0
        for i in range(len(key) - 1, -1, -1):
            buff += key[i] * radix ** i
            pass
        
        key = buff
        return key % self.size
    
    def find(self, key, shift=1):
        for i in range(self.size):
            if self.table.get(self.hash_key(key) + i*shift) == key:
                return self.hash_key(key) + i*shift
        
        return None
    
    def insert(self, key, shift=1):
        for i in range(self.size):
            if not self.table.get(self.hash_key(key) + i*shift):
                self.table.update({self.hash_key(key) + i*shift : key})
                return self.hash_key(key) + i*shift
            pass
        
        raise IndexError
        pass
    
    def delete(self, key, shift=1):
        for i in range(self.size):
            if self.table.get(self.hash_key(key) + i*shift) == key:
                self.table.update({self.hash_key(key) + i*shift : None})
                return self.hash_key(key) + i*shift
        
        raise IndexError
        pass
    
    def class_name(self):
        return 'Remainder_Division'
```

```python
class Hash_XOR(Hash_Remainder_Division):
    
    def hash_key(self, key):
        
        buff = 0
        for i in range(len(key)):
            buff ^= ord(key[i])
        
        return buff
    
    def class_name(self):
        return 'XOR'
```

```python
class Hash_Direct_Address(Hash_Table):
    
    def __init__(self, data_dict=None, size=1000):
        super().__init__(size)
        self.keys = {v:k for k, v in data_dict.items()}
        pass
    
    def find(self, key):
        return self.table[key]
    
    def insert(self, data):
        self.table[self.keys.get(data)] = data
        pass
    
    def delete(self, data):
        self.table[self.keys.get(data)] = None
        pass
    
    def class_name(self):
        return 'Direct_Address'
```

```python
class Hash_Multitable(Hash_Table):
    
    def __init__(self, size=1000, step=10):
        self.multitable = []
        self.size = size
        for i in range(size):
            self.multitable.append({i:None for i in range(size / step)})
            pass
        pass
```

```python
def check_hashing_time(hash_table, data, mode):
    
    times = []
    
    counter = 0
    for i in data:
        
        counter += 1
        if counter > hash_table.size:
            break
        
        if mode == 'find':
        
            timer = time.process_time()

            hash_table.find(i)

            timer = time.process_time() - timer
            times.append(timer)    
            pass
        
        elif mode == 'insert':
            
            timer = time.process_time()

            hash_table.insert(i)

            timer = time.process_time() - timer
            times.append(timer)
            pass
        
        elif mode == 'delete':
            
            timer = time.process_time()

            hash_table.delete(i)

            timer = time.process_time() - timer
            times.append(timer)
            pass
        
        pass
    
    return times
```

```python
def research_time(hasher, data_frame, testing_data_dict, mode):
    
    check_time = check_hashing_time(hasher, list(testing_data_dict.values()), mode)

    data_frame[hasher.class_name() + '_' + mode] = pd.Series(np.array(check_time))

    return data_frame
```

```python
testing_data = 0

with open('test.txt', 'r', encoding='utf-8') as f:
    testing_data = f.readlines()[0].split()
    pass
f.close()

testing_data_dict = {}
for i in range(len(testing_data)):
    testing_data_dict.update({i : testing_data[i]})
    pass
```

```python
hash_tables_size = 9000
DF = pd.DataFrame()
```

```python
%%time

hasher_hda = Hash_Direct_Address(testing_data_dict, size=hash_tables_size)

research_time(hasher_hda, DF, testing_data_dict, 'insert')
research_time(hasher_hda, DF, testing_data_dict, 'delete')

del hasher_hda
```
